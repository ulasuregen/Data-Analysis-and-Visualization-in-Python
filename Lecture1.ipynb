# Basics

```python
my_list = list(range(10))
print(my_list)
```

```python
print( my_list[-1 ] )
print( my_list[1:5] )
print( my_list[3:]  )
```

To get every *kth* element of the list, use a[::k]. This operation can be combined with taking an
interval.

```python
print( my_list[::2]  )
print( my_list[1:7:2] )
```

## len( )
Use len to get length of the list

```python
print( len( my_list ) )
print( len( my_list[::2] ) )
```

## Copy Lists & Arrays when reassigning !!

```python
a = [1, 2, 3]
b = a
a[0] = 5

print(b)
```

```python
a = [1, 2, 3]
b = a[:]
a[0] = 24

print(b)
```

# Numpy NP
NumPy arrays are fixed-type, N-dimensional arrays that are the standard numeric array type in Python scientific code

```python
import numpy as np
```

### np.array & array.shape

```python
arr = np.array([1,2,3])
mat = np.array([[1,2,3], [4,5,6]])

print(arr.shape)
print(mat.shape)
```

### array.dtype

```python
mat.dtype
```

#### astype
You change the datatype by astype

```python
mat = mat.astype(np.float64)
mat.dtype
```

## + and * operators

* Use + to sum element-wise two arrays
* Use * with a number to multiply each element 

```python
unsorted = np.array([3, 1, 4, 2])

5 * unsorted, unsorted + unsorted
```

## sort - sum - max - argmax  

```python
np.sort(unsorted), np.sum(unsorted)
```

```python
np.argmax(unsorted), np.max(unsorted)
```

## Randomness

### Random Integers

np.random.randint(low, high, size)

```python
np.random.seed(0)

print("randInts:", np.random.randint(low= 0, high=10, size = 5))
# low inclusive
# high exclusive
```

### Normal Distribution
np.random.normal(loc, scale, size)

```python
print("Normal Distribution:", np.random.normal(loc = 0.0, scale = 1.0, size = 5))
```

### Uniform

np.random.rand

```python
print("Uniform: ", np.random.rand(5))
```

### Choice with probabilities
np.random.choice

```python
print("choice:", np.random.choice(["red", "green", "blue"], size = 5, p = [0.1, 0.7, 0.2]))
```

# Exercise

```python
from plotnine import ggplot, aes, geom_line, theme_minimal, theme
```

```python
def parabola(n):
    x = np.array(range(-n, n+1))
    y = x**2
    df = pd.DataFrame({'x': x, 'y':y})

    plot = ( ggplot(df, aes(x = 'x', y = 'y'))  
           + geom_line() 
        #   + theme_minimal() 
           + theme(figure_size= (4,4))
    )

    return plot

parabola(10)
```
